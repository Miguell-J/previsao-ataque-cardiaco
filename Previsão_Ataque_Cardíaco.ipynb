## Instalando as ferramentas necessárias

```python
import pandas as pd
import sweetviz as sv
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

## Descrição dos dados

- Age : Idade do paciente

- Sex : Sexo do paciente

- exang: angina induzida por exercício (1 = sim; 0 = não)

- CA: Número de grandes navios (0-3)

- cp : Tipo de dor torácica tipo dor torácica tipo

 - Valor 1: angina típica
 - Valor 2: angina atípica
 - Valor 3: dor não anginosa
 - Valor 4: assintomático

- trtbps : pressão arterial de repouso (em mm Hg)

- chol : colo em mg/dl obtido via sensor de IMC

- Sesclerose Feminina : (glicemia de jejum > 120 mg/dL) (1 = verdadeiro; 0 = falso)

- rest_ecg : resultados eletrocardiográficos de repouso

 - Valor 0: normal
 - Valor 1: apresentar anormalidade da onda ST-T (inversões da onda T e/ou supradesnivelamento ou depressão do segmento ST de > 0,05 mV)
 - Valor 2: hipertrofia ventricular esquerda provável ou definida pelos critérios de Estes
- thalach : frequência cardíaca máxima atingida

- output: 0= menos chance de ataque cardíaco 1= mais chance de ataque cardíaco

## Pré processamento dos dados

```python
x = data.drop("output", axis=1)
y = data["output"]

x
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=42)

x_teste
```

```python
scaler = MinMaxScaler()
x_treino_n = scaler.fit_transform(x_treino)
x_teste_n = scaler.fit_transform(x_teste)

x_teste_n
```

## Criação do modelo de floresta de decisão

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

clf = RandomForestClassifier()
model_rf = clf.fit(x_treino_n, y_treino)

preds = model_rf.predict(x_teste_n)

accuracy_rf = accuracy_score(y_teste, preds)
```

## Avaliando o modelo

```python
print("Acertividade do modelo de RandomForest {:.2%}".format(accuracy_rf))
```

```python
import matplotlib.pyplot as plt

plt.scatter(range(len(y_teste)), y_teste, label='Real')
plt.scatter(range(len(preds)), preds, label='Previsto')
plt.title('Análise das previsões')
plt.xlabel('Amostras')
plt.ylabel('Classe')
plt.legend(loc='upper right')
plt.show()
```
